## Import Libraries

```python
import pandas as pd
pd.options.display.max_rows = 999
pd.options.display.max_columns = 999
```

## Data Massage

```python
df = pd.read_csv('reviews_staten_island.csv')
```

```python
df.head()
```

```python
df = df.drop(columns=['Unnamed: 0'], axis=1)
```

```python
df['count'] = df.groupby('Restaurant_Id').cumcount() + 1
```

```python
df = df.set_index(['Restaurant_Id'])
```

```python
df.head()
```

## Pivot DataFrame to long form

```python
pivot_table = df.pivot_table(index=['Restaurant_Id'],
                               columns=['count'],
                             values=['Review'],
                             aggfunc=lambda x: ' '.join(str(v) for v in x))
```

```python
pivot_table.head()
```

## Rename and adjust Columns

```python
columns_name = ['review_'+str(i+1) for i in range(149)]
```

```python
pivot_table.columns = columns_name
```

```python
pivot_table = pivot_table.reset_index()
```

```python
pivot_table.head()
```

```python
pivot_table.to_csv('fulltablestatenisland.csv')
```
