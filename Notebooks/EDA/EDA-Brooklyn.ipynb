```python
import pandas as pd
pd.options.display.max_rows = 999
pd.options.display.max_columns = 999
```

```python
df = pd.read_csv('reviews2.csv')
```

```python
df.head()
```

```python
df = df.drop(columns=['Unnamed: 0'], axis=1)
```

```python
df['count'] = df.groupby('Restaurant_Id').cumcount() + 1
```

```python
df = df.set_index(['Restaurant_Id'])
```

```python
df.head()
```

```python
pivot_table = df.pivot_table(index=['Restaurant_Id'],
                               columns=['count'],
                             values=['Review'],
                             aggfunc=lambda x: ' '.join(str(v) for v in x))
```

```python
pivot_table.head()
```

```python
columns_name = ['review_'+str(i+1) for i in range(138)]
```

```python
pivot_table.columns = columns_name
```

```python
pivot_table = pivot_table.reset_index()
```

```python
pivot_table.head()
```

```python
pivot_table.to_csv('fulltablebrkln.csv')
```
