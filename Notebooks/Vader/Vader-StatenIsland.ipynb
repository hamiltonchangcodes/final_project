```python
import pandas as pd 
import matplotlib as plt
import seaborn as sns
import nltk
from vaderSentiment import vaderSentiment as vader 
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from tqdm import tqdm
pd.options.display.max_columns = 999
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
vad = ['review_vader_'+ str(x+1) for x in range(149)]
lis = ['features'+str(x+1) for x in range(149)]
nums = [x for x in range(1,150)]
rva = ['review_'+ str(x+1) for x in range(149)]
```

```python
df = pd.read_csv('../../Data/Fullsets/cleanedStatenIslandfull.csv')
```

```python
df.head()
```

```python
df = df.fillna(value=str('none'))
```

```python
df.head(15)
```

```python
for j in tqdm(nums):
    lis[j-1] = df[rva[j-1]].apply(lambda x: analyzer.polarity_scores(x)).apply(pd.Series)
    df = pd.concat([df,lis[j-1].apply(pd.Series)],1)
    df.rename(columns={'neg': 'neg'+str(j), 'neu': 'neu'+str(j), 'pos': 'pos'+str(j), 'compound': 'compound'+str(j)}, inplace=True)
    
```

```python
df.to_csv('vaderStatenIsland.csv')
```
