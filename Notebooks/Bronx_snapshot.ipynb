```python
#create list of dictionaries with 3 positive, 3 negative
#for review+j find compound+j 
#if compound+j >.5
#then append to posreview1
# make a counter to stop at 3
#make a list of positive reviews, then dictionary entries will be index position of list.

countpos = 0
countneg = 0
nums = [str(s+1) for s in range(260)]
snapshots =[]
for row in vd[['restaurant_name']].itertuples():
    posreview = []
    negreview = []
    #iterate through nums using variable j
    for j in nums:
        #iterate if counts positive is less than 3
        if not np.isnan(df.iloc[row[0]]['compound'+j]):
            if countpos < 3:
                #if the score in compound number J is greater than .5
                if vd.iloc[row[0]]['compound'+j]>.5:
                #append to posreview list from review number j
                    posreview.append(vd.iloc[row[0]]['review_'+j])
                #add to 1 to counts
                    countpos+=1
            #iterate if counts positive is less than 3
            if countneg < 3: 
                #if compound score is less than .5
                if vd.iloc[row[0]]['compound'+j]<.5:
                    #append to negreview list the corresponding review
                    negreview.append(vd.iloc[row[0]]['review_'+j])
                #add to 1 to counts
                    countneg+=1
    #rest counts    
    countpos = 0
    countneg = 0 
    #add to dictionary the following positions.    
    snapshots.append({'restaurant': row[1],
                      'Positive Review 1': posreview[0],
                      'Positive Review 2': posreview[1],
                      'Positive Review 3': posreview[2],
                      'Negative Review 1': negreview[0],
                      'Negative Review 2': negreview[1],
                      'Negative Review 3': negreview[2]})
```
