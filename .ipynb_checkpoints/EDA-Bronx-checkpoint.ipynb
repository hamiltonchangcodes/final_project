```python
df.rename(columns={'0':'restaurant_id', '1':'Review'}, inplace=True)
```

```python
df1.rename(columns={'0':'restaurant_id', '1':'Review'}, inplace=True)
```

```python
df3.rename(columns={'0':'restaurant_id', '1':'Review'}, inplace=True)
```

```python
df.head()
```

```python
df1.head()
```

```python
df = df.drop(columns=['Unnamed: 0'], axis=1)
```

```python
df1 = df1.drop(columns=['Unnamed: 0'], axis=1)
```

```python
truedf = pd.concat([df, df1, df3])
```

```python
truedf.head()
```

```python
truedf['count'] = truedf.groupby('restaurant_id').cumcount() + 1
```

```python
truedf.head(150)
```

```python
truedf = truedf.set_index(['restaurant_id'])
```

```python
df.head()
```

```python
pivot_table = truedf.pivot_table(index=['restaurant_id'],
                               columns=['count'],
                             values=['Review'],
                             aggfunc=lambda x: ' '.join(str(v) for v in x))
```

```python
pivot_table.head()
```

```python
columns_name = ['review_'+str(i+1) for i in range(260)]
```

```python
pivot_table.columns = columns_name
```

```python
pivot_table = pivot_table.reset_index()
```

```python
pivot_table.head()
```

```python
pivot_table.to_csv('fulltablebronx.csv')
```
