```python
df = pd.read_csv('reviews_manhattan.csv')
```

```python
import string 
def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text
```

```python
df1 = df[df['Review'].str.contains('service')]
```

```python
df1 = df1.fillna(value=str('none'))
```

```python
df1['Review'] = df1['Review'].apply(remove_punctuations)
```

```python

df1['Review'] = df1['Review'].apply(lambda x: ' '.join([word.lower() for word in x.split() if word.lower() not in (stop)]))
```

```python

df1['Review'] = df1['Review'].apply(lambda x: ' '.join([lemmatizer.lemmatize(word, pos = 'v') for word in x.split()]))
```

```python
df1.drop(columns=['Unnamed: 0', 'Restaurant_Id'], axis=1, inplace=True)
```

```python
df1['token_Review']= df1['Review'].apply(tokenizer)
```

```python
df1.drop(columns=['Review'], axis=1, inplace=True)
```

```python
tokenslist = df1.values.tolist()
```

```python
temp = []
for row in tokenslist:
    if type(row) == list:
        temp.extend(row)
```
